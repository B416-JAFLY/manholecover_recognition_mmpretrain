```python
import json
import matplotlib.pyplot as plt

# scalars.json
data = []
with open('/home/u2021213565/jupyterlab/manholecover_recognition_mmpretrain/mmpretrain/work_dirs/resnet50_1xb32_funct_pretrain/20240713_161739/vis_data/scalars.json', 'r') as f:
    for line in f:
        data.append(json.loads(line))

# 
steps_loss = []
losses = []
steps_acc = []
accuracies = []

for entry in data:
    if 'loss' in entry:
        steps_loss.append(entry['step'])
        losses.append(entry['loss'])
    if 'accuracy/top1' in entry:
        steps_acc.append(entry['step'])
        accuracies.append(entry['accuracy/top1'])

# 
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(steps_loss, losses, label='Loss', color='red')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend()

# 
plt.subplot(1, 2, 2)
plt.plot(steps_acc, accuracies, label='Accuracy', color='blue')
plt.xlabel('Step')
plt.ylabel('Accuracy')
plt.title('Training Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
